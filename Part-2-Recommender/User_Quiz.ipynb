Welcome to our national park recommender! Many national parks offer a wide range of experiencese, so we understand that it can be hard to choose which park to visit next. Just answer a couple of simple questions to let us know what you prefer, and we'll recommend the perfect parks for you.

Let's go!

```python
import recs
```

```python
try:
    activities, itineraries, learnings, landscapes, weathers, seasons, length, popularity = recs.prompts()
    areas = ['all']
    trails = recs.get_trails_in_area(areas)
    parks = recs.get_parks_from_trails(trails)
    tfidfs = [recs.vec(c, parks) for c in ['description','activities', 'weatherInfo']]
    sum_tfidf = recs.sum_tfidfs(tfidfs, parks)
    inputs = [landscapes,
seasons,
weathers,
learnings,
itineraries,]
    recommended = [recs.recommend(s, sum_tfidf) for s in inputs]
    merged_recommendations = recs.merge_sum(recommended)
    recs.display_recs(parks, merged_recommendations)
    print("\n---\nAlso, based on your interets on hiking trails, here are the hiking trials we think you will like:")
    users_top_trails = recs.create_ranking(activities, length, popularity)
    display(users_top_trails)
except:
    print("We could not find a park match for you. Please try again. We recommend selecting no more than two responses per question.")
```
