```python
import requests
import wikipedia
from bs4 import BeautifulSoup
import requests
from bs4 import BeautifulSoup,SoupStrainer
import decimal
import time
import os.path
from os import path
import csv
state_park_dict={}
 
state_list =[]
park_list=[]
park_link=''
page = requests.get('https://en.wikipedia.org/wiki/Lists_of_state_parks_by_U.S._state')
soup = BeautifulSoup(page.content, 'html.parser')
info = soup.find_all('li')
two_decimal_places = decimal.Decimal("0.01")
NoCoordinates = ['California Citrus State Historic Park','Indio Hills Palms']
count = 0
count1 = 0

state_parks_page = []
for li in info:
    state_link = li.a.get('href')
    time.sleep(3)
    state_park_page = requests.get("https://en.wikipedia.org" + state_link)
    state_parks_page.append(state_park_page)
    print('state_line=%s' % state_link)
```

```python
dcnt = 0
skip_index = [6,12,14,16,17,20,21,22,27,28,29,31,32,37,38,39,41,50,51,52,53]
parks_url_dict = {}
for count, page in enumerate(state_parks_page):
    if count in skip_index:
        continue
    soup = BeautifulSoup(page.content, 'html.parser')
    if count not in parks_url_dict.keys():
        parks_url_dict[count] = []
    for tr in soup.find_all('tr'):
        data = tr.find('td')
        if data != None:
            if 'title' in data.contents[0].attrs.keys() and 'page does not exist' not in data.contents[0]['title']:
                if "href" in data.contents[0].attrs.keys():
                    park_link = data.contents[0]['href']
                    if park_link!= None:
                        park_url = "https://en.wikipedia.org"+park_link
                        print(park_url)
                        time.sleep(1)
                        parkresponse = requests.get("https://en.wikipedia.org"+park_link)
                        parks_url_dict[count].append(parkresponse)
```

```python
parks_link_dict = {}
for count, page in enumerate(state_parks_page):
    if count in skip_index:
        continue
    soup = BeautifulSoup(page.content, 'html.parser')
    if count not in parks_link_dict.keys():
        parks_link_dict[count] = []
    for tr in soup.find_all('tr'):
        data = tr.find('td')
        if data != None:
            if 'title' in data.contents[0].attrs.keys() and 'page does not exist' not in data.contents[0]['title']:
                if "href" in data.contents[0].attrs.keys():
                    park_link = data.contents[0]['href']
                    parks_link_dict[count].append(park_link)
```
